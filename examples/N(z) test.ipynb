## The basics:

- `get_available_catalogs()` lists available catlaogs; returns `dict`.
- `load_catalog()` loads the catalog you want; returns an instance of `GCR.BaseGenericCatalog`

```python
import GCRCatalogs

## check version
print('GCRCatalogs =', GCRCatalogs.__version__, '|' ,'GCR =', GCRCatalogs.GCR.__version__)
```

```python
## list all find available catlaogs, sorted by their name

print('\n'.join(sorted(GCRCatalogs.get_available_catalogs())))
```

## Plot N(z)

```python
for catalog in ('buzzard_test', 'ezmock_elg'):
    gc = GCRCatalogs.load_catalog(catalog)
    z = gc.get_quantities(['redshift'], filters=['is_ELG'])['redshift']
    plt.hist(z, 50, alpha=0.4, density=True, label=catalog);
    
plt.xlabel('$z$')
plt.ylabel('$N(z)$ (normalized)')
plt.title('ELG')
plt.legend()
```
